# Modify players data

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
```

```python
del players["Rk"]
```

```python
# remove the * at the end of some player's name
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

```python
players.head()
```

```python
# problem: some players has played for multiple teams and posittion -> we only want the TOT stat

# example: group by a player and a year
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))

# how to solve: make sure every group only have one row -> every player should only have one row of stat in each year 
```

```python
# before applying the fucntion single_team
players[players["Player"] == "Greg Anderson"]
```

```python
def single_team(df):
    # if the group only has one row (the player only played for one team on that year) -> do nothing
    if df.shape[0]==1:
        return df
    else:
        # get the row with the column "TM"="TOT"
        row = df[df["Tm"]=="TOT"]
        # replace "TOT" with the last row's "TM" -> the player at that year will be playing for only the last team he was in with the total stat 
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
# apply the function single_team to each group that is group by Player and Year
players = players.groupby(["Player", "Year"]).apply(single_team)

```

```python
players.head(20)
```

```python
# reomve the left most column which is a duplicate
players.index = players.index.droplevel()
```

```python
# reomve the left most column which is a duplicate
players.index = players.index.droplevel()
```

# Merge mvp with players

```python
# outer merge - players in the player dataframe that are not found in the mvp dataframe will remain after merging 
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
# problem: we want to change NaN to 0 for the last three column
combined
```

```python
# solution: apply fillna to the df -> assign the new df to the original df
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
# solved
combined
```

# Modify team

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(10)
# problem 1 : there are rows with Division names
# prblem 2: there are rows with team name containing "*" at the end
```

```python
# solution 1: for any row that "W" contains "Division" will be removed 
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
# solution 2: remove the "*" at the team column
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
# solved
teams.head(10)
```

```python
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
# problem: if we want to merge "combined" with "teams", the team column is different
```

```python
# create a dictionary where the keys are the abbrev while the values are the team's full name
nicknames = {}
with open("nicknames.csv") as f:
    # lines will be a list of lines read from the csv file
    lines = f.readlines()
    for line in lines[1:]:
        # first remove the "\n", then split by the "," and assgin them to the respective value
        abbrev,name = line.replace("\n","").split(",")
        # store the key and value pair into the dictionary
        nicknames[abbrev] = name

```

```python
# first use the map method to convert the abbrev in "Tm" column to the team's full name, then assign the column to "Team"
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
# merge "teams" with "combined" to "train"
train = combined.merge(teams, how="outer",on=["Team", "Year"])
```

```python
train
```

```python
del train["Unnamed: 0"]
```

```python
# check the data types of each data 
train.dtypes
# problem: some columns such as Age should be numerical values instead of object type
```

```python
# solution: go through every column in "train" and turn it to numerical values, do nothing if error is encountered
train = train.apply(pd.to_numeric, errors='ignore')
```

```python
# solved
train.dtypes
# problem: "GB" is not converted to numerical values as expected
```

```python
# check what is the possible issue -> there is a "-" within the values
train["GB"].unique()
```

```python
# solution: replace the "-" with "0"
train["GB"] = pd.to_numeric(train["GB"].str.replace("—","0"))
```

```python
# solved
train.dtypes
```

```python
train.to_csv("player_mvp_stats.csv")
```
